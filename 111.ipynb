# Q1

```python
import numpy as np

# 1. 使用arange创建1到12的一维数组
arr = np.arange(1, 13)  # 注意arange是左闭右开，所以结束值写13

# 2. reshape成3行4列的二维数组
arr_2d = arr.reshape(3, 4)

# 3. 分别打印维度(ndim)、形状(shape)、元素总个数(size)
print("数组的维度(ndim):", arr_2d.ndim)
print("数组的形状(shape):", arr_2d.shape)
print("数组的元素总个数(size):", arr_2d.size)
```

# Q2

```python
import numpy as np

# 创建包含浮点数的NumPy数组
float_arr = np.array([1.5, 2.6, 3.7, 4.8])
print("转换前的数组：", float_arr)
print("转换前的数据类型名称：", float_arr.dtype.name)

# 使用astype转换为32位整数
int_arr = float_arr.astype(np.int32)
print("转换后的数组：", int_arr)
print("转换后的数据类型名称：", int_arr.dtype.name)
```

# Q4

```python
import numpy as np

# 给定的一维数组
arr = np.array([10, 15, 20, 25, 30, 35])

# 1. 花式索引提取第1、3、5个元素（索引为0,2,4）
fancy_index = arr[[0, 2, 4]]
print("花式索引结果：", fancy_index)

# 2. 布尔索引提取所有大于20的元素
bool_index = arr[arr > 20]
print("布尔索引结果：", bool_index)
```

# Q6

```python
import numpy as np

# 给定数组
x = np.array([1, 4, 9, 16])
y = np.array([2, 1, 4, 5])

# 1. 计算x中每个元素的平方根
sqrt_x = np.sqrt(x)
print("x的平方根数组：", sqrt_x)

# 2. 结合两个数组，取对应位置的最大值
result = np.maximum(sqrt_x, y)
print("对应位置的最大值：", result)
```

```python
import numpy as np

# 创建形状为(2, 3, 4)的三维数组
arr = np.arange(24).reshape((2, 3, 4))

# 使用transpose()方法转换轴顺序为(1, 0, 2)
transposed_arr = arr.transpose((1, 0, 2))

# 打印转换后新数组的形状
print(transposed_arr.shape)
```

# Q9

```python
import numpy as np

# 创建二维数组
arr = np.array([[6, 2, 7], [3, 6, 2], [4, 3, 2]])

# 1. 计算所有元素的算术平均值
mean_val = arr.mean()
print("算术平均值：", mean_val)

# 2. 沿 axis=0（列方向）排序
arr.sort(axis=0)
print("排序后的数组：")
print(arr)
```

# Q10

```python
import numpy as np

# 1. 设置随机数种子
np.random.seed(10)

# 2. 生成两个形状为(2, 2)的随机矩阵A和B
A = np.random.rand(2, 2)
B = np.random.rand(2, 2)

# 3. 计算矩阵A和B的点积
dot_product = np.dot(A, B)

# （可选）打印结果验证
print("矩阵A：")
print(A)
print("矩阵B：")
print(B)
print("矩阵点积结果：")
print(dot_product)
```
